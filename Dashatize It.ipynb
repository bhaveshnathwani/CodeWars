**Description:**
<br>
Given a number, return a string with dash'-'marks before and after each odd integer, but do not begin or end the string with a dash mark.
<br>
<br>
**Example:**
<br>
dashatize(274) -> '2-7-4'
<br>
dashatize(6815) -> '68-1-5'



```python
def dashatize(num):
    
    if isinstance(num, int):

        #Convert number to list for manipulation
        num_list = list(map(int, str(num))) 

        num_list2 = []
        for i in num_list[0:]:
            if i == 0:
                num_list2.append(str(i))
            elif i%2==0 and i!=0:
                num_list2.append(str(i))
            elif i%2!=0 and i!=0:
                num_list2.append("-")
                num_list2.append(str(i))
                num_list2.append("-")

        #Join the list of numbers and dashes back into a string
        numbers = "".join(num_list2)

        #Check for more than 1 dash at any point
        indexes_to_delete = []
        for i in range(1,(len(numbers))):
            if numbers[i] == "-" and numbers[i-1] == "-":
                indexes_to_delete.extend([i])

        #Kick out indexes where a double dash has been identified
        formatted_list = [i for j, i in enumerate(numbers) if j not in indexes_to_delete]
        formatted_string = "".join(formatted_list)

        #Remove leading and trialing dashes
        if formatted_string[0] == "-":
            formatted_string = formatted_string[1:]

        if formatted_string[-1] == "-":
            formatted_string = formatted_string[0:-1]

        return formatted_string
    else:
        return "None"
```

### <font color=#C70039>Testing with random integers:</font> 

```python
import random

test_list = [random.randrange(0,100000000000000) for i in range(500)]

for random_number in test_list:
    print("{} returns {}".format(random_number, dashatize(random_number)))
```
