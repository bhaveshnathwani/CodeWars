### The Challenge:
https://www.codewars.com/kata/simple-pig-latin/train/python
<br>
<br>
**Description:**
Move the first letter of each word to the end of it, then add "ay" to the end of the word. 
<br>
Leave punctuation marks untouched.
<br>
<br>
**Examples:**
<br>
pig_it('Pig latin is cool') # igPay atinlay siay oolcay
<br>
pig_it('Hello world !')     # elloHay orldway !

```python
def pig_it(text):
    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    piglatin = []
    for i in text.split():
        if i[-1].upper() not in alphabet:
            character = i[-1]
        else:
            character = ""
        if len(i) == 1 and i.upper() not in alphabet:
            piglatin.append(i)
        else:
            piglatin.append(str(i[1:]+i[0]+"ay"+str(character)))
    return " ".join(piglatin)
```

### Testing with random strings:

```python
import random
nouns = ("puppy", "car", "rabbit", "girl", "monkey")
verbs = ("runs", "hits", "jumps", "drives", "barfs") 
adv = ("crazily", "dutifully", "foolishly", "merrily", "occasionally")
adj = ("adorable", "clueless", "dirty", "odd", "stupid")
sentences = []
for i in range(500):
    num = random.randrange(0,5)
    sentences.append(nouns[num] + ' ' + verbs[num] + ' ' + adv[num] + ' ' + adj[num])

for i in sentences:
    print("{} returns: {}".format(i, pig_it(i)))
```
