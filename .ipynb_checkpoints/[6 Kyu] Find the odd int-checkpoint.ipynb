**Description:**
Given an array, find the int that appears an odd number of times.
<br>
<br>
**NOTE:** There will always be only one integer that appears an odd number of times.

```python
def find_it(seq):
    return [i for i in seq if seq.count(i)%2 == 1][0]
```

### Testing with random list of integars, such that there is one element that appears an odd number of times:

```python
from collections import Counter

def test_random_list(listsize, lowestnum, highestnum):
    
    random_list = [random.randint(lowestnum,highestnum) for i in range(listsize)]
    random_list = [val for val in random_list for _ in (0, 1)]

    #Add in a random element once so that it appears in the list an odd number of times
    odd_element = random.randint(lowestnum,highestnum)
    random_list.extend([odd_element])

    for number in range(len(random_list)):
        odd_occurances = []
        if random_list.count(random_list[number])%2 != 0 and random_list[number] not in odd_occurances :
            odd_occurances.extend([random_list[number]])

    if len(odd_occurances)>0:
        for i in odd_occurances:
            print("{} Appears {} times".format(i,random_list.count(i)))
    return random_list
```

```python
for i in range(100):
    print(test_random_list(10,0,50))
```
