**Description:**
Complete the method/function so that it converts dash/underscore delimited words into camel casing. 
<br>
The first word within the output should be capitalized only if the original word was capitalized
<br>
(known as Upper Camel Case, also often referred to as Pascal case).
<br>
<br>
**Examples:**
<br>
to_camel_case("the-stealth-warrior") # returns "theStealthWarrior"
<br>
to_camel_case("The_Stealth_Warrior") # returns "TheStealthWarrior"

```python
import re

def to_camel_case(text):
    if len(text) >0:
        delimiter = [delimiter for delimiter in text if delimiter.upper() not in "ABCDEFGHIJKLMNOPQRSTUVQXYZ"][0]
        original_list = re.split('[^a-zA-Z]', text)
        new_list = list(map(lambda x:x.title(), re.split('[^a-zA-Z]', text)))
        if len(text)>0:
            if new_list[0][0].lower() == original_list[0][0]:
                new_list[0] = new_list[0][0].lower()+new_list[0][1:]
        return "".join(new_list)
    else:
        return ""
```

### Testing with random strings:

```python
import random
nouns = ("puppy", "car", "rabbit", "girl", "monkey")
verbs = ("runs", "hits", "jumps", "drives", "barfs") 
adv = ("crazily", "dutifully", "foolishly", "merrily", "occasionally")
adj = ("adorable", "clueless", "dirty", "odd", "stupid")
delimiter = ("-", ",", "_", "++", "@@")
sentences = []
for i in range(500):
    num = random.randrange(0,5)
    sentences.append(nouns[num] + delimiter[num] + verbs[num] + delimiter[num] + adv[num] + delimiter[num] + adj[num])

for i in sentences:
    print("{} returns: {}".format(i, to_camel_case(i)))
```
