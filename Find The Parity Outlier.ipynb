**Description:**
You are given an array (which will have a length of at least 3, but could be very large) containing integers.  
<br>
The array is either entirely comprised of odd integers or entirely comprised of even integers except for a single integer N.
<br>
Write a method that takes the array as an argument and returns this "outlier" N.
<br>
<br>
**Examples:**
<br>
[2, 4, 0, 100, 4, 11, 2602, 36]
<br>
Should return: 11 (the only odd number)
<br>
<br>
[160, 3, 1719, 19, 11, 13, -21]
<br>
Should return: 160 (the only even number)

```python
def find_outlier(integers):
    even = [int for int in integers if int%2 == 0] 
    odd = [int for int in integers if int%2 != 0]
    return  even[0] if len(even) == 1 else odd[0] 
```

```python
#A one liner solution, although difficult to follow, would be the following:
def oneline_find_outlier(integars):
    return  [int for int in integers if int%2 == 0][0] if len([int for int in integers if int%2 == 0]) == 1 else [int for int in integers if int%2 != 0][0]
```

### <font color=#C70039>Testing with random even integars and 1 odd:</font> 
**Note** We expect all returned output to be even

```python
import random

def even_list_test():
    
    even_numbers = []
    odd_numbers = []
    for i in range(random.randint(3,10)):
        even_numbers.append(random.randint(1,49)*2)

    odd_numbers.append(random.randint(1,49)*2 +1)

    return even_numbers + odd_numbers
```

```python
even_test = []
for i in range(100):
    even_test.append(even_list_test())
    
for i in range(100):
    print("{} returns {}".format(even_list_test(), find_outlier(even_list_test())))
```

### <font color=#C70039>Testing with random odd integars and 1 even:</font> 
**Note** We expect all returned output to be even

```python
import random

def odd_list_test():
    
    even_numbers = []
    odd_numbers = []
    for i in range(random.randint(3,10)):
        odd_numbers.append(random.randint(1,49)*2+1)

    even_numbers.append(random.randint(1,49)*2)

    return odd_numbers + even_numbers
```

```python
odd_test = []
for i in range(100):
    odd_test.append(odd_list_test())
    
for i in range(100):
    print("{} returns {}".format(odd_list_test(), find_outlier(odd_list_test())))
```
