<br>
<br>
**Description:**
<br>
Given a number, return a string with dash'-'marks before and after each odd integer, but do not begin or end the string with a dash mark.
<br>
<br>
**Example:**
<br>
dashatize(274) -> '2-7-4'
<br>
dashatize(6815) -> '68-1-5'



```python
def dashatize(num):
    if isinstance(num, int):
        number = abs(num)
    
        numlist = []
        
        for i in str(number):
            if int(i) == 0 or i == "-":
                numlist.append(i)
            elif int(i) % 2 == 0:
                numlist.append(i)
            elif int(i) % 2 != 0:
                numlist.extend(["-", str(i), "-"])
        
        if numlist[0] == "-":
            numlist = numlist[1:]
        if numlist[-1] == "-":
            numlist = numlist[:-1]
        
        output = "".join(numlist)
        
        while "--" in output:
            output = output.replace("--", "-")
        
        return output
        
    else:
        return 'None'
    
```

### Testing with random integers:

```python
import random

test_list = [random.randrange(0,100000000000000) for i in range(500)]

for random_number in test_list:
    print("{} returns {}".format(random_number, dashatize(random_number)))
```
